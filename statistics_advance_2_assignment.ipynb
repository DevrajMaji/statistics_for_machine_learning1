Question 1: What is hypothesis testing in statistics?
  ->Hypothesis testing in statistics is a method used to make decisions about a population based on sample data.

It works like this:

We start with an assumption (null hypothesis(H₀)) about the population (ex:- the mean of the population is 50).

Then, we collect sample data and calculate a test statistic.

We compare this test statistic with a critical value or use a p-value to decide whether the sample data provides enough evidence to reject the null hypothesis.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
->null hypothesis (H₀):

It is the default assumption that there is no effect, no difference, or no relationship in the population.

ex:The average height of students = 170 cm.

alternative hypothesis (Ha):

It is what we want to prove. It states that there is an effect, a difference, or a relationship.

ex: The average height of students ≠ 170 cm.

differences:
null hypothisis(H₀):-
Assumes no effect or no difference.
Purpose is basis for testing.
Symbol is H₀.
ex:-Mean = 170

alternative hypothesis (H₁):-
Assumes there is an effect or difference.
Purpose is competes against H₀.
Sysmbol is H₁ or Ha.
ex:-Mean ≠ 170.


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
->The significance level is the threshold probability used to decide whether to reject the null hypothesis.

It represents the risk of making a Type I error (rejecting H₀ when it is actually true).

Role in Hypothesis Testing:

Set before the test begins. ex: α = 0.05 means we allow a 5% chance of wrongly rejecting H₀.

The p-value (calculated from sample data) is compared with α:

If p ≤ α → Reject H₀ (evidence supports H₁).

If p > α → Fail to reject H₀ (not enough evidence against H₀).

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
  -> Z-Test

Used when population variance (σ²) is known.

Suitable for large sample size (n > 30).

Based on the standard normal distribution.

Distribution shape is fixed (does not depend on sample size).

ex: Testing if average marks of 1000 students = 60, with known σ.

T-Test

Used when population variance (σ²) is unknown.

Suitable for small sample size (n ≤ 30).

Based on the Student’s t-distribution.

Distribution shape depends on degrees of freedom (n−1), becomes normal as n grows.

ex: Testing if average marks of 15 students = 60, with unknown σ.

When to Use Each

Z-Test:

When population variance is known.

When sample size is large (n > 30).

ex: Testing if the average height of 1000 students = 170 cm, where σ is known.

T-Test:

When population variance is unknown (uses sample standard deviation).

When sample size is small (n ≤ 30).

ex: Testing if the average height of 15 students = 170 cm, where σ is unknown.

```python
"""Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]"""

from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu_0 = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
alpha = 0.05
print("sample Mean:", sample_mean)
print("sample Std Dev:", sample_std)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)
if p_value < alpha:
    print("reject Null Hypothesis (H0). there is a significant difference.")
else:
    print("fail to Reject Null Hypothesis (H0). no significant difference.")
```

```python
"""Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
"""

from scipy import stats
np.random.seed(42)
mu, sigma = 50, 5
n = 100
data = np.random.normal(mu, sigma, n)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
alpha = 0.05
se = sample_std / np.sqrt(n)
ci = stats.t.interval(1 - alpha, df=n-1, loc=sample_mean, scale=se)
print("Sample Mean:", sample_mean)
print("95% CI:", ci)
plt.hist(data, bins=15, color="skyblue", edgecolor="black", density=True)
plt.axvline(sample_mean, color="red", linestyle="--", label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci[0], color="green", linestyle="--", label=f"95% CI Lower = {ci[0]:.2f}")
plt.axvline(ci[1], color="green", linestyle="--", label=f"95% CI Upper = {ci[1]:.2f}")
plt.title("nor mal distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
"""Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
"""

def zscore_analysis(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    print("First 10 Z-scores:", z_scores[:10])
    plt.hist(z_scores, bins=15, color="skyblue", edgecolor="black", density=True)
    plt.axvline(0, color="red", linestyle="--", label="Mean (0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
    return z_scores
data = np.array([49, 50, 51, 52, 48, 47, 53, 50, 49, 51])
z_scores = zscore_analysis(data)
```
